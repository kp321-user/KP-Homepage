```python
import feedparser


def main():
    feed_url = "https://www.reddit.com/r/python/.rss"
    feed = feedparser.parse(feed_url)

    for entry in feed.entries:
        print(f"Title: {entry.title}")
        print(f"Link: {entry.link}")
        print(f"Published: {entry.published}")
        print("-" * 40)


if __name__ == "__main__":
    main()
```

```python
from xml.etree import ElementTree as ET
import requests
import re
from urllib.parse import urlparse

# url = "https://raw.githubusercontent.com/plenaryapp/awesome-rss-feeds/refs/heads/master/countries/without_category/Canada.opml"
url = "https://raw.githubusercontent.com/plenaryapp/awesome-rss-feeds/refs/heads/master/countries/without_category/Australia.opml"

response = requests.get(url)
response.raise_for_status()

with open("australia_feeds.opml", "w", encoding="utf-8") as f:
    f.write(response.text)
cleaned = re.sub(r"&(?!amp;|lt;|gt;|quot;|apos;|#\d+;)", "&amp;", response.text)
# Replace & with &amp; unless it's already a valid entity like &amp; &lt; &gt; &quot; &apos;
with open("australia_feeds.opml", "w", encoding="utf-8") as f:
    f.write(cleaned)

tree = ET.parse("australia_feeds.opml")  # reads the file, builds the tree
root = tree.getroot()  # the top node — here, the <opml> tag
for outline in root.iter("outline"):
    title = outline.get("title", "").strip()
    rss_url = outline.get("xmlUrl")
    if not title:
        title = urlparse(rss_url).netloc.replace("www.", "")

    print(f"{title:50}  {rss_url}")
```

```python
import os
from dotenv import load_dotenv
import requests
from datetime import datetime, timedelta

load_dotenv()

query = str(input("Enter a search query: "))
from_date = (datetime.now() - timedelta(days=25)).strftime("%Y-%m-%d")
to_date = datetime.now().strftime("%Y-%m-%d")
sort_by = "relevancy"
client = os.getenv("NEWS_API_KEY")

params = {
    "q": query,
    "sortBy": sort_by,
    "apiKey": client,
}
response = requests.get("https://newsapi.org/v2/everything", params=params)
data = response.json()

if data.get("status") != "ok":
    print("Error:", data.get("message"))
else:
    for article in data["articles"]:
        print(f"Title: {article['title']}")
        print(f"Description: {article['description']}")
        print("-" * 40)
```
